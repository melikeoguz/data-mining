```python
def get_variable_type(element) :
    """
     Sütunların sürekli veya kategorik değişken olup olmadığını kontrol edin.
     Varsayım şudur:
                  benzersiz sayı <20 ise kategorik olduğunu varsayıyoruz
                  benzersiz sayı> = 20 ve dtype = [int64 veya float64] ise sürekli olduğunu varsayıyoruz
     """
    if element==0:
        return "Not Known"
    elif element < 20 and element!=0 :
        return "Categorical"
    elif element >= 20 and element!=0 :
        return "Contineous"
    
def predict_variable_type(metadata_matrix):
    metadata_matrix["Variable_Type"] = metadata_matrix["Valeurs_Uniques_Count"].apply(get_variable_type).astype(str)
    metadata_matrix["frequency"] = metadata_matrix["Null_Count"] - metadata_matrix["Null_Count"]
    metadata_matrix["frequency"].astype(int)
    return metadata_matrix 

def get_meta_data(dataframe) :
    """
     İletilen herhangi bir veri kümesinde meta veri elde etme yöntemi
    """
    metadata_matrix = pd.DataFrame({
                    'Datatype' : dataframe.dtypes.astype(str), # types de données de colonnes
                    'Non_Null_Count': dataframe.count(axis = 0).astype(int), # nombre total d'éléments dans les colonnes
                    'Null_Count': dataframe.isnull().sum().astype(int), # total des valeurs nulles dans les colonnes
                    'Null_Percentage': dataframe.isnull().sum()/len(dataframe) * 100, # pourcentage de valeurs nulles
                    'Valeurs_Uniques_Count': dataframe.nunique().astype(int) # nombre de valeurs uniques
                     })
    
    metadata_matrix = predict_variable_type(metadata_matrix)
    return metadata_matrix

def list_potential_categorical_type(dataframe,data) :
    print("*********potansiyel kategorik veri türü sütunları*********")
    metadata_matrix_categorical = dataframe[dataframe["Variable_Type"] == "Categorical"]
    
    length = len(metadata_matrix_categorical)
    if length == 0 :
        print("Belirli bir veri kümesinde kategorik sütun yok.")  
    else :    
        metadata_matrix_categorical = metadata_matrix_categorical.filter(["Datatype","Valeurs_Uniques_Count"])
        metadata_matrix_categorical.sort_values(["Valeurs_Uniques_Count"], axis=0,ascending=False, inplace=True)
        col_to_check = metadata_matrix_categorical.index.tolist()
        name_list = []
        values_list = []
        
        for name in col_to_check :
            name_list.append(name)
            values_list.append(data[name].unique())
        
        temp = pd.DataFrame({"index":name_list,"Valeurs_Uniques":values_list})
        metadata_matrix_categorical = metadata_matrix_categorical.reset_index()
        metadata_matrix_categorical = pd.merge(metadata_matrix_categorical,temp,how='inner',on='index')
        display(metadata_matrix_categorical.set_index("index"))
```

```python
metadata = get_meta_data(data)

#List potential columns of categorical variables
list_potential_categorical_type(metadata,data)
```

<h2><code>Multiple Linear Regression Nedir ?</code></h2>

Çoklu regresyon olarak da bilinen çoklu doğrusal regresyon (MLR), bir yanıt değişkeninin sonucunu tahmin etmek için birkaç açıklayıcı değişken kullanan istatistiksel bir tekniktir.
Çoklu doğrusal regresyonun (MLR) amacı, açıklayıcı (bağımsız) değişkenler ile yanıt (bağımlı) değişken arasındaki doğrusal ilişkiyi modellemektir.

Tahminler ve değerlendirmeler sonucunda verinin ne kadar tutarlı olduğu saptanılmaktadır.

<p> Çoklu doğrusal regresyon gerçekleştirmek için aşağıdaki varsayımların karşılanması gerekir:</p>

<h4>--- Model yapımından önce: ---</h4>

<p>Doğrusal ilişki: Bağımlı değişken Y (yani Fiyat), bağımsız değişkenler X ile doğrusal bir ilişkiye sahiptir ve bunu doğrulamak için XY dağılım grafiğinin doğrusal olduğundan emin olunmalıdır.</p>

<b>Çoklu Doğrusallık:</b><p> Çoklu regresyon, bağımsız değişkenler X'in birbiriyle güçlü bir şekilde ilişkili olmadığını varsayar. Bu varsayım, Varyans Enflasyon Faktörü (VIF) veya Korelasyon Matrisi kullanılarak test edilir.</p>

<h4>Modelin Hata Analizi ---</h4>
<ul>                

<li>Hata Dağılımının Normalliği</li>

<h3 style="text-align:center">Bağımsız Değişkenlerin Görselleştirilmesi</h3>

```python
import scipy
from scipy.stats.stats import pearsonr

def pairplot(x_axis,y_axis) :
    sns.pairplot(data,x_vars=x_axis,y_vars=y_axis,height=4,aspect=1,kind="scatter")
    plt.show()
```

```python
y_vars=['price']
```

    **Korelasyon katsayısı** negatif ise iki değişken arasında **ters ilişki** vardır. Korelasyon katsayısı pozitif ise değişkenler arasında doğru orantı bulunmaktadır. Eğer korelasyon katsayısı 0.8> ise çok yüksek korelasyon olduğu yorumu yapılır.

<h3><code>1) Fiyat vs Dingil mesafesi - Boş ağırlık - Boreratio</code></h3>

```python
x_vars=['wheelbase','curbweight','boreratio']
pairplot(x_vars,y_vars)
print("İlk bakışta, 3 değişken pozitif olarak ilişkilidir, ancak daha yüksek değerlerde yayılır.")

p1=data['wheelbase']
p2=data['curbweight']
p3=data['boreratio']

pearson_coeff, p_value = pearsonr(p1,data['price'])
print('\nKorelasyon Katsayısına bakarak bundan emin olabiliriz')
print('\nFiyat ve dingil mesafesi arasındaki Korelasyon Katsayısı:',pearson_coeff*100,'%')

pearson_coeff, p_value = pearsonr(p2,data['price'])
print('Fiyat ve boş ağırlık arasındaki korelasyon katsayısı:',pearson_coeff*100,'%')

pearson_coeff, p_value = pearsonr(p3,data['price'])
print('Price ve boreratio arasındaki korelasyon katsayısı: ',pearson_coeff*100,'%')
```

<h3><code>1-Anderson-Darling Testi</code></h3>

**Anderson-Darling Testi**, istatistik bilim dalında, bir parametrik olmayan istatistik sınaması olup örneklem verilerinin belirli bir olasılık dağılımı gösterip göstermediğini sınamak için, yani uygunluğun testi için kullanılmaktadır. Ayrıca hataların standardını kontrol etmek için uzmanlar tarafından en çok kullanılan yöntemlerden biridir.

Aşağıdaki verileri kontrol eder.

**H0:** Veriler normal dağılımı takip eder

**H1:** Veriler normal dağılımı takip etmiyor

H0 nasıl kontrol edilir?

Tarafımızdan kritik bir değerle ilişkilendirilen bir önem düzeyi (bizim durumumuzda% 5) seçilir.

İade edilen A-
D istatistiği, seçilen anlamlılık düzeyi için kritik değerden büyükse, H0 reddedilmelidir.

```python
from scipy import stats

def normality_of_residuals_test(model):
    '''
   Hataların QQ grafiğini oluşturma ve hataların normalliğini incelemek için Anderson-Darming istatistiksel testini gerçekleştirme 
    
    Arg:
    * model - OLS models adapted from statsmodels
    '''
    sm.ProbPlot(lm.resid).qqplot(line='s');
    plt.title('Q-Q plot');

    ad = stats.anderson(lm.resid, dist='norm')
    
    print(f'----Anderson-Darling test ---- \nstatistic: {ad.statistic:.4f}, critical value of 5%: {ad.critical_values[2]:.4f}')
    
normality_of_residuals_test(lm)
```

İz bir yay şekline sahip değil.

test H0'ın verilerin normal dağılımı takip ettiğini kabul eder (istatistikler <kritik değer% 5)

Hataların aslında bağımlı olduğunu söyleyen bir eğilimin varlığıdır.

```python
plt.figure(figsize=(15,9))
plt.scatter(resid.index, resid.values)
plt.hlines(0,0,200)
plt.suptitle('Siparişlere Göre Hatalar', fontsize=16)
plt.xlabel('Sipariş')
plt.ylabel('Hatalar')
```

Bu sayının 2 civarında çıkması, **otokorelasyon vardır** boş hipotezini reddedemeyeceğimizi gösterir. Artık terimlerin birbirine bağımlı (otokorelasyonlu) olmadığının hipotezini test eder.

Test, 0 ile 4 arasında bir değer verir.

**D = 2 değeri:** Örnekte hiçbir otokorelasyon saptanmadı.

<h3><code>2-Homoscedasticity</code></h3>

Bu varsayım, artıkların varyansının bağımsız değişkenlerin değerleri için benzer olması gerektiğini gösterir.

Kalıntıları tahmin edilen değerlere göre çizerek bunu doğrulayabiliriz.

Grafikte homo-dedastisiteyi belirlemek için noktaların konumu rastgele olmalı ve hiçbir eğilim görünür olmamalı,
ve grafikteki kırmızı regresyon çizgisi olabildiğince düz olmalıdır (kavis şeklinde olmamalıdır).

Goldfeld-Quandt testi:

Şunları test eder:

**H0:** Hata terimleri homoskedastiktir

**Alternatif Hipotez H1:** Hata terimleri heteroskedastiktir.

P değeri <= 0.05 ==> ise sıfır hipotezini reddediyoruz H0

P-Değeri> 0 ise.
05 ==> H0 boş hipotezi doğrudur

```python
%matplotlib inline
%config InlineBackend.figure_format ='retina'
import seaborn as sns 
import matplotlib.pyplot as plt
import statsmodels.stats.api as sms
from statsmodels.compat import lzip
sns.set_style('darkgrid')
sns.mpl.rcParams['figure.figsize'] = (15.0, 9.0)

def homoscedasticity_test(model):
    '''
    Linear regresyon modelinde artık eş varyans testi işlevi.

    Hataları, tahmin edilen değerlerle karşılaştırır ve Goldfeld-Quandt testlerini çalıştırır.
    
    Args:
    * model - fitted OLS model from statsmodels
    '''
    fitted_vals = model.predict()
    resids = model.resid

    #fit_reg=False
    sns.regplot(x=fitted_vals, y=resids, lowess=True, line_kws={'color': 'red'})
    plt.suptitle('Hatalar ve Tahminler', fontsize=16)
    plt.xlabel('Tahminler')
    plt.ylabel('Hatalar')

    print('\n----Goldfeld-Quandt test ----')
    name = ['F statistic', 'p-value']
    test = sms.het_goldfeldquandt(lm.resid, lm.model.exog)
    print(lzip(name, test))
    print('\n----Residuals plots ----')

homoscedasticity_test(lm)
```

**Analiz:**

P-Değeri testi Goldfeld-Quandt> 0.05, bu nedenle H0'ın hata terimlerinin homoskedastik olduğunu, yani hataların sabit bir varyansa sahip olduğunu söylemeyi kabul ediyoruz.

Ayrıca, yukarıda gördüğümüz noktalar rastgele dağılmış, bulunma eğilimi yok ve çizgi bir kemer şeklinde değil

tüm hipotezler doğrulandı

<h2><code>RFE (Recursive Feature Elimination) Nedir ?</code></h2>

Yinelemeli Özellik Kaldırma veya kısaca RFE, popüler bir özellik seçim algoritmasıdır.

RFE'nin yapılandırması ve kullanımı oldukça kolaydır. Ayrıca bir veri setinde hedef değişkeni tahmin etmede daha çok veya en alakalı olan özellikleri (sütunları) seçmede etkilidir.
RFE'yi kullanırken iki önemli yapılandırma seçeneği vardır: 

<ul>
<li>seçilecek özellik sayısının seçimi</li>
<li>özelliklerin seçilmesine yardımcı olmak için kullanılan algoritmanın seçimi</li>
</ul>

Bu hiperparametrelerin her ikisi de araştırılabilir, ancak yöntemin performansı bu hiperparametrelerin iyi yapılandırılmasına büyük ölçüde bağlı değildir.

```python
import statsmodels.api as sm 

def color_code_vif_values(val):
    """
    Take a scalar and return a string with the property css 'color: red' for 10, black otherwise.
    """
    if val > 10 : color = 'red' 
    elif val > 5 and val <= 10 : color = 'blue'
    elif val > 0 and val <= 5 : color = 'darkgreen'
    else : color = 'black'
    return 'color: %s' % color

def drop_col(dataframe,col_to_drop) :
    dataframe.drop([col_to_drop],axis=1,inplace=True)
    return dataframe

def display_vif(x) :
    #Calculer les VIFs pour le nouveau modèle
    from statsmodels.stats.outliers_influence import variance_inflation_factor
    vif = pd.DataFrame()
    X = x
    vif['Features'] = X.columns
    vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    vif['VIF'] = round(vif['VIF'], 2)
    vif = vif.set_index("Features")
    vif = vif.sort_values(by = "VIF", ascending = False)
    df = pd.DataFrame(vif.VIF).style.applymap(color_code_vif_values)
    display(df)
    
model_count = 0

def statsmodel_summary(y_var,x_var) :
    global model_count
    model_count = model_count + 1
    text = "*****MODEL - " + str(model_count)
    print(text)
    
    x_var_const = sm.add_constant(x_var) # adding constant
    lm = sm.OLS(y_var,x_var_const).fit() # calculating the fit
    print(lm.summary()) # print summary for analysis
    display_vif(x_var_const.drop(['const'],axis=1))
    return x_var_const , lm
```

        P-value > **0.05** gelişme yoktur

**2.VIF (Variance Inflation Factor):**

        VIF > **5**, multicollinearitenin mevcut olabileceğine dair bir işarettir. Ancak bağımsız değişkenlerinizi yeniden gözden geçirip silinecek kadar değildir.

        VIF > **10**, değişkenler arasında çoklu bağlantı vardır.
